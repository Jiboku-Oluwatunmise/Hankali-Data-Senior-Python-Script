```python
student['Tot'] = student['Physics_Marks']+ student['Math_Marks'] + student['Chemistry_Marks']

student
```

```python
student['Tot'] = ((student[['Physics_Marks','Math_Marks','Chemistry_Marks']].sum(axis=1)))

student
```

```python
student['Total']= (student['Tot']/300)*100

student
```

```python
# Delete a column

student.drop('Tot', axis=1, inplace=True)

student
```

```python
# create a function to create labels for Age

def Age_Group(Age):
  if Age < 25:
    return 'Young'
  elif Age <= 35:
    return 'Adult'
  elif Age <= 45:
    return 'Middle Aged'
  else:
    return 'Aged'

student['Age_Group'] = student['Age'].apply(Age_Group)

student
```

```python
# create a function to create labels for IQ

def iq_range(IQ):
  if IQ < 100:
    return 'Low'
  elif IQ <= 120:
    return 'Average'
  elif IQ <= 140:
    return 'High'
  else:
    return 'Gifted'

student['IQ_Range'] = student['IQ'].apply(iq_range)

student
```

```python
# Count of SHG

student['Study_Hours_Group'].value_counts()
```

```python
# convert Study_Hours_Group to numeric

student['SHG'] = student['Study_Hours_Group'].map({'Low': 1, 'High': 2})

student.head(5)
```

```python
# convert hours & Total studied to 2 dp

student['Hours_Studied'] = student['Hours_Studied'].round(2)
student['Total'] = student['Total'].round(2)

student.head(5)
```

```python
#Describe the numerical data

student.describe()
```

```python
# Filter students based on study hours group

df1 = student.set_index('Study_Hours_Group')
df1
```

```python
# Present df with high study hours

df1.filter(like='High', axis=0)
```

```python
# Sorting values in a dataframe based on column

df1[df1['IQ'] >= 100].sort_values(by=['Total','Age_Group'],ascending=[False,True])
```

# Data Visualisation with Seaborn and **Matplotlip**

Seaborn and Matplotlip are python libaries used to present data visualisations in python. They are both relatively similar, however, seaborn is used for making statistical graphics in python while matplotlib is used for creating static, and interactive visualizations in Python. Both helps curate visualisations for the data, however use case depends on the context.

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Plot the correlation as a visualisation with Seaborn

sns.heatmap(student3.corr(), annot = True)

plt.rcParams['figure.figsize'] = (20,20)

plt.show()
```

```python
# Get the outliers for the Totals using pandas

student3['Total'].plot.box(figsize=(5,5))
```

```python
# Get the outliers for the IQ using pandas

student3['IQ'].plot.box(figsize=(7,5))
```

```python
# Get the scatter plot for two variables using pandas

student.plot.scatter(x = 'IQ', y = 'Total', s = 10, c = 'Red')
```

```python
# Get the scatter plot for two variables using pandas

student3.plot.scatter(x = 'Total', y = 'Hours_Studied', s = 10, c = 'Green')
```

```python
# Get the histogram using matplotlip and pandas
plt.figure(figsize=(2,5))

student3['IQ'].plot.hist(bins = 5)

plt.show()
```

```python
# A graph comparing the age group with the IQ range using seaborn

plt.figure(figsize=(15,8))

sns.histplot(data=student, x="IQ_Range", hue="Age_Group", multiple="stack")

plt.show()
```

```python
# Plot a seaborn visualisation comparing Age with Total

a = sns.relplot(data=df1, x="Age", y="Total", col="IQ_Range")
a.set(ylim=(0, 100))
```

```python
sns.pairplot(data=student3, hue="IQ_Range")
```

```python
# make pie chart using matplotlib

x = student['IQ_Range'].value_counts()

plt.figure(figsize=(8, 6))
plt.pie(x, labels=x.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Student IQ Ranges')
plt.axis('equal')

plt.show()
```

```python
# Find out who studied more between male and feemale

plt.figure(figsize=(10,7))

ax = sns.countplot(x='Study_Hours_Group', data=student, hue='Gender')

plt.title('Hours Studied between Male and Female')
plt.xlabel('Study Hours')
plt.ylabel('Count of Grouped Study Hours ')

for i in ax.containers:
    ax.bar_label(i,)

sns.set_theme(style="whitegrid", palette="pastel")
```
